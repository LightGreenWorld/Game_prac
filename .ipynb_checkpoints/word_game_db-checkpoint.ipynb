```python
import os
import numpy as np
import pandas as pd
from selenium import webdriver
```

```python
driver = webdriver.Chrome('--')
```

```python
url='--'
driver.get(url)
# driver.maximize_window()
```

```python
xpath = "//table[@class='prettytable']//dd"
el_list = driver.find_elements_by_xpath(xpath)
```

```python
el_list2 = []

for el in el_list:
    voca = el.text
    voca = voca.split(" ")[0]
    el_list2.append(voca)
```

```python
el_list2 = list(set(el_list2)) # 중복제거
print(el_list2)
```

```python
el_list3 = []
for el in el_list2:
    if len(el) !=1 :
        el_list3.append(el)
el_list3.sort()                  # 오름차순 정렬
print(el_list3)
```

```python
# txt 파일로 저장하는 방법
f = open("--", 'w')
for el in el_list3:
    f.write(el+' ')
f.close()
```
